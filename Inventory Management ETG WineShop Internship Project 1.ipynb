```python
def updateDatabase(data):
    data1=json.dumps(data)
    fd=open("myWineShop.json","w")
    fd.write(data1)
    fd.close()
    
```

```python
def addItemToInventory(data,productID):
    productID=str(productID)
    fd=open("myWineShop.json","r")
    data3=fd.read()
    fd.close()
    data3=json.loads(data3)
    
    for i in data3:
        if i==productID: 
            print("Product ID already exists")
            return
    
    data3[productID]=data
    
    updateDatabase(data3)
    
    print("Item added to the inventory successfully")
    
    
    
```

```python
def calculateTotalSales():
    fd=open("myWineShop.json","r")
    data4=fd.read()
    fd.close()
    data4=json.loads(data4)
    totalSales=0
    for i in data4:
        totalSales+=data4[i]["price"]*data4[i]["amountPurchased"]
    return totalSales
        
```
